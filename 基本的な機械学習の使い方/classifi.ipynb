# ライブラリのインポート

```python
from sklearn.linear_model import LogisticRegression as LR
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,classification_report
from sklearn.preprocessing import StandardScaler,MinMaxScaler
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# データの読み込みと確認

```python
df=pd.read_csv("breast_cancer.csv")
df.head()
```

# 説明変数(X)と目的変数(Y)に分ける

```python
y_name="y"
y=df[y_name].values
x_table=df.drop([y_name],axis=1)
x_name=x_table.columns
x=x_table.values
```

# 訓練データとテストデータに分ける(正確にやるなら検証用データも作る)

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.3)
```

# 機械学習モデルの生成と学習

```python
model=LR()
model.fit(x_train,y_train)
```

# テストデータによる予測

```python
y_pred=model.predict(x_test)
```

# 混合行列

```python
con=pd.DataFrame(confusion_matrix(y_test,y_pred))
con.index=["陰性","陽性"]
con.columns=["陰性","陽性"]
con
```

# 各精度指標

```python
print("accuracy      =%.2f"%(accuracy_score(y_test,y_pred)))
print("recall    nega=%.2f posi=%.2f"%(recall_score(y_test,y_pred,average=None)[0],recall_score(y_test,y_pred,average=None)[1]))
print("presicion nega=%.2f posi=%.2f"%(precision_score(y_test,y_pred,average=None)[0],precision_score(y_test,y_pred,average=None)[1]))
print("F1score   nega=%.2f posi=%.2f"%(f1_score(y_test,y_pred,average=None)[0],f1_score(y_test,y_pred,average=None)[1]))
```

```python
print(classification_report(y_test,y_pred))
```
