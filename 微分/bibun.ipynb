# 微分
微分では導関数を作成して極値を算出することで値の増え方・減り方などを推定できる

# ライブラリのインポート
最新版にすると数式をきれいに表示してくれる。

```python
import sympy as sy
import matplotlib.pyplot as plt
```

# シグモイド関数の作成
シグモイド関数は機械学習のとりわけロジスティック回帰に使われるだけでなくニューラルネットワークでは微分値も使われるため例として使用する。

```python
x=sy.symbols("x")
y=1/(1+sy.exp(-x))
```

```python
y
```

```python
sy.plot(y)
```

# シグモイド関数の一回微分
シグモイド関数を一回微分した関数で初期のニューラルネットワークでは誤差の修正を行う。この導関数で誤差を計算する上で最大値が0.25となることを確認する。

```python
sy.diff(y,x)
```

```python
sy.plot(sy.diff(y,x))
```

```python
sy.diff(y,x).subs(x,0)
```

# シグモイド関数の二回微分

```python
sy.diff(y,x,x)
```

```python
sy.plot(sy.diff(y,x,x))
```

# 極値の算出
一回微分では関数の増減が分かり、二回微分では関数の増え方(減り方)が分かる。

```python
y=x**3+6*x**2+9*x+12
y
```

一回微分することで極値を取り増減の変化をするx座標を算出する

```python
dydx=sy.diff(y)
dydx
```

```python
sy.solve(dydx)
```

二回微分することで増減の仕方(上に凸か下に凸か)の変化をするx座標の位置を算出する

```python
d2ydx2=sy.diff(y,x,x)
d2ydx2
```

```python
sy.solve(d2ydx2)
```

元の数式のy座標が0の時のx座標の地点を把握する

```python
sy.solve(y)
```

元の数式のx座標が0の時のy座標の地点を把握する

```python
y.subs(x,0)
```

```python
sy.plot(y)
```

一回微分したときの数式を因数分解

```python
sy.factor(dydx)
```

二回微分した時の数式を展開

```python
sy.expand(d2ydx2)
```

# 多変数の微分

```python
y=sy.symbols("y")
z=sy.sin(x)*sy.cos(y)
z
```

xとyで偏微分を行う

yで偏微分

```python
dzdy=sy.diff(z,y)
dzdy
```

xで偏微分

```python
dzdx=sy.diff(z,x)
dzdx
```

```python
dx=sy.symbols("dx")
dy=sy.symbols("dy")
```

全微分では各変数で偏微分した値に偏微分した時の変数のd_(xの場合dx)を纏めて全て足す

```python
(dzdy)*(dy)+(dzdx)*(dx)
```
